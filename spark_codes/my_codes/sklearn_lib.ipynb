```python
import numpy as np
import sklearn.datasets as load_iris
import sklearn.linear_model as Perceptron

iris = load_iris.load_iris()
print(iris)
X = iris.data[:,(2,3)]
print(X)
y = (iris.target == 0).astype(int)

per_clf = Perceptron.Perceptron()
per_clf.fit(X,y)
print(per_clf)

y_pred = per_clf.predict([[2,0.5]])
print("Output : ", y_pred)

```
