```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['Mumbai', 'Delhi', 'Bangalore']}

df = pd.DataFrame(data)

print(df)
display(df)
```

```python
df['Capital City'] = df['City'].apply(lambda x: x.upper() if type(x) == str else x)
print(df)
```

```python
df['New Age'] = df['Age'].apply(lambda x: x + 5 if x >= 30 else x)
print(df)
```

```python
df.loc[df['Age'] >= 30, 'New Age2'] = df['Age'] + 5
print(df)
```

```python
df.iloc[[0][0], 'New Age2'] = df['Age'] + 5
print(df)
```

```python
import pandas as pd
import numpy as np

# Creating a default DataFrame
df = pd.DataFrame({
    'A': range(1, 7),
    'B': range(10, 16),
    'C': range(20, 26)
})

print(df)
# help(df.apply)
df['A'] = df.apply(lambda x: x['A'] + 1000 if x['A']%2==0 else 'False', axis=1)
df['B'] = df['B'].apply(np.square)
print(df)
```

```python
import pandas as pd
from datetime import datetime

data = {
    "id": list(range(1, 21)),
    "empid": list(range(101, 121)),
    "department": [
        "HR", "IT", "Finance", "Marketing", "IT",
        "HR", "Finance", "IT", "Marketing", "HR",
        "IT", "Finance", "Marketing", "HR", "Finance",
        "IT", "Marketing", "HR", "Finance", "IT"
    ],
    "salary": [
        50000, 70000, 60000, 55000, 72000,
        52000, 63000, 71000, 57000, 53000,
        74000, 65000, 58000, 54000, 66000,
        73000, 59000, 56000, 67000, 75000
    ],
    "joining_date": [
        datetime(2020, 5, 1), datetime(2019, 6, 15), datetime(2021, 3, 10),
        datetime(2018, 7, 25), datetime(2022, 1, 5), datetime(2020, 8, 14),
        datetime(2019, 9, 30), datetime(2021, 12, 20), datetime(2017, 4, 5),
        datetime(2023, 2, 15), datetime(2016, 11, 3), datetime(2020, 1, 22),
        datetime(2018, 6, 7), datetime(2022, 9, 10), datetime(2019, 5, 6),
        datetime(2021, 10, 13), datetime(2023, 3, 18), datetime(2020, 7, 9),
        datetime(2019, 8, 23), datetime(2021, 11, 30)
    ],
    "last_date": [
        datetime(2025, 5, 1), datetime(2024, 6, 15), None, datetime(2026, 7, 25), None,
        None, datetime(2023, 9, 30), datetime(2025, 12, 20), datetime(2024, 4, 5), None,
        datetime(2026, 11, 3), None, datetime(2027, 6, 7), datetime(2025, 9, 10), None,
        datetime(2026, 10, 13), None, datetime(2024, 7, 9), None, None
    ]
}

df = pd.DataFrame(data)
print(df)
```

```python
df1 = df.copy()
df2 = df1.groupby('department').agg(total_sal = ('salary', 'sum'), jd = ('joining_date', 'min')).reset_index()
print(df2)
df3 = df1.groupby('department').agg(total_sal = ('salary', 'min')).reset_index()
print(df3)
df4 = df1.groupby('department').agg(min_sal = ('salary', 'max')).reset_index()
print(df4)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'invoice_date': pd.to_datetime(['2024-01-01', '2024-01-05', '2024-02-10', '2024-01-02', '2024-03-15']),
    'invoice_amount': [100, 200, 150, 300, 250],
    'cust_id': [1, 1, 2, 2, 1]
}

df = pd.DataFrame(data)

# Aggregation
result_df = df.groupby('cust_id').agg(
    first_invoice_date=('invoice_date', 'min'),
    last_invoice_date=('invoice_date', 'max'),
    total_invoice_amount=('invoice_amount', 'sum')
).reset_index()

print(result_df)
```
