# Groupby

```python
import pandas as pd
```

```python
oo = pd.read_csv('../data/olympics.csv',skiprows=4)
oo.head()
```

```python
oo.groupby(['Event_gender', 'Medal','NOC']).count()
```

```python
oo[oo['NOC']== 'IND']
```

```python
oo.groupby(['Event_gender', 'Medal','NOC']).count().loc[(slice(None), slice(None), 'IND'), :]
```

```python
indoo = oo[oo.NOC == 'IND']
indoo.groupby(['Event_gender', 'Medal','NOC']).count()
```

### The Groupby object

```python
oo.groupby('Edition')
```

```python
type(oo.groupby('Edition'))
```

```python
list(oo.groupby('Edition'))
```

## Iterating through groups

```python
for group_key, group_value in oo.groupby('Edition'):
    print(group_key)
    print(group_value)
```

```python
type(group_value)
```

## Groupby computations

```python
oo.groupby('Edition').size()
```

### agg([...])

```python
oo.groupby(['Edition','NOC','Medal']).agg(['min','max','count'])
```

```python
oo.groupby(['Edition','NOC','Medal']).size()
```

### agg({  : [ ... ]})

```python
oo.groupby(['Edition','NOC','Medal']).agg({'Edition' :['min','max','count']})
```

```python
oo.loc[oo.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition' : ['min','max','count']})
```
