## Plot types

### What were the different sports in the first olympics? Plot them using different graphs.

```python
fo = oo[oo.Edition == 1896]
fo.head()
```

### Line plot

```python
plt.figure(figsize=(9, 2))  # Adjust width and height as needed
fo.Sport.value_counts().plot(kind='line', marker='o')
plt.xticks(rotation=45);  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--')  # Adds dashed gridlines on the x-axis
```

```python
plt.figure(figsize=(18, 3))  # Adjust width and height as needed
fo.Sport.value_counts().plot(kind='line', marker='o')
plt.grid(axis='both', linestyle='--')  # Adds dashed gridlines on the x-axis
```
