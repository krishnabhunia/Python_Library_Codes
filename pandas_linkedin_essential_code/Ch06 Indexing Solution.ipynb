# Solution

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oo = pd.read_csv('../data/olympics.csv',skiprows=4)
oo.head()
```

### Plot the total number of medals awarded at each of the Olympic games throughout history.

```python
plt.figure(figsize=(10,5))
oo.Edition.value_counts().sort_index().plot(marker='x', color='red')
plt.grid(axis='both', linestyle='--', linewidth=3, color='green')
plt.xticks(color='purple')
plt.yticks(color='brown');
```

```python
plt.figure(figsize=(10,5))
oo.Edition.value_counts().sort_index().plot(marker='x', color='red')
plt.grid(axis='both', linestyle='--', linewidth=3, color='green')
plt.xticks(color='purple')
plt.yticks(color='brown')
plt.xlabel("Edition", color='Green');
plt.ylabel("Number of medals", color='purple');
```
